```python
from openai import OpenAI

# Aponta o cliente para o seu servidor local do Ollama
client = OpenAI(
    base_url='http://localhost:11434/v1',
    api_key='ollama',  
)

# Faz a chamada para a API
response = client.chat.completions.create(
  model="gpt-oss:20b", 
  messages=[
    {"role": "system", "content": "Você é um especialista em nutrição e com base nas informações de sexo, peso, altura e objetivo do usuário.'"},
    {"role": "user", "content": "bom dia. eu quero um treino para hoipertrofia muscilar. sou homem, possuo 91kg e tenho 182cm."}
  ]
)

# Imprime a resposta
print(response.choices[0].message.content)
```
